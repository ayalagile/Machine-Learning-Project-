```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder

df = pd.read_csv("train_u6lujuX_CVtuZ9i.csv")

df.drop("Loan_ID", axis=1, inplace=True)

df["Dependents"] = df["Dependents"].replace("3+", 3).astype(float)

num_cols = ["ApplicantIncome", "CoapplicantIncome", "LoanAmount", "Loan_Amount_Term", "Credit_History", "Dependents"]
cat_cols = ["Gender", "Married", "Education", "Self_Employed", "Property_Area"]

df[num_cols] = df[num_cols].fillna(df[num_cols].mean())
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])

df["Loan_Status"] = df["Loan_Status"].map({"Y":1, "N":0})

binary_cols = ["Gender", "Married", "Education", "Self_Employed"]
le = LabelEncoder()
for col in binary_cols:
    df[col] = le.fit_transform(df[col])

df = pd.get_dummies(df, columns=["Property_Area"], drop_first=True)

num_features = ["ApplicantIncome", "CoapplicantIncome", "LoanAmount", "Loan_Amount_Term", "Credit_History", "Dependents"]
scaler = StandardScaler()
df[num_features] = scaler.fit_transform(df[num_features])

print(df.isnull().sum())
print(df.head())
```

```python
X = df.drop("Loan_Status", axis=1)
y = df["Loan_Status"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
y_pred_log = log_model.predict(X_test)
```

```python
from sklearn.svm import SVC

svm_model = SVC(kernel='linear', probability=True)
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

models = {
    "Logistic Regression": y_pred_log,
    "SVM": y_pred_svm,
    "kNN": y_pred_knn
}

for name, y_pred in models.items():
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    incorrect = (y_test != y_pred).sum()
    print(f"{name}: Incorrect={incorrect}, Accuracy={acc:.2f}, Precision={prec:.2f}, Recall={rec:.2f}")
```
